```python
def psnr(img, img_noisy, max_pixel):
    mse = np.mean((img - img_noisy) ** 2)
    if(mse == 0): 
        return 100
    psnr = 20 * math.log10(max_pixel / math.sqrt(mse))
    return psnr
```

```python
def f(lamb,b):
    return np.exp(-1* (np.power(lamb,2))/(np.power(b,2)))
```

```python
def aniso(img, steps, b, lamb = 0.25): 
    
    img_new = np.zeros(img.shape) 
    for t in range(steps): 
        dx = img[:-2,1:-1] - img[1:-1,1:-1] 
        dy = img[2:,1:-1] - img[1:-1,1:-1] 
        dz = img[1:-1,2:] - img[1:-1,1:-1] 
        dw = img[1:-1,:-2] - img[1:-1,1:-1] 
        img_new[1:-1,1:-1] = img[1:-1,1:-1] +lamb * (f(dx,b)*dx + f (dy,b)*dy + f (dz,b)*dz + f (dw,b)*dw) 
        img = img_new 
    return img
```

```python
img = imageio.volread('/home/haas/Downloads/cameraman.tif') 

im_min, im_max = img.min(), img.max()
img = (img - im_min) /(float)(im_max - im_min)  
img_noisy = img + np.random.normal(0,0.2,size=(img.shape))

pyp.figure('Image noisy')
pyp.imshow(img_noisy, cmap='gray')
pyp.axis('on')

pyp.show()
psnr_img = psnr(img,img_noisy,1.)
print(f'PSNR:',psnr_img)
  

img2 = aniso(img_noisy, 20, 0.5, 0.15)
pyp.figure('Image smooth')
pyp.imshow(img2, cmap='gray')
pyp.axis('on')

pyp.show()
psnr_img = psnr(img,img2,1.)
print(f'PSNR:',psnr_img)
```
